```python
df=mongo_saver.get_data('coinmarketcap_spot_downloaded_1min')
df['quote'] = df['quote'].apply(lambda x: x['USD'])
df = pd.concat([df[df.columns.difference(['quote'])], pd.json_normalize(df.quote)], axis=1)
df
```

```python
# Looking for anomalies in the data
# Checking that the cryptocoin prices are positive
price = (data['price']>=0).all()
# Checking that the volumes traded is positive
volume_24h = (data['volume_24h']>=0).all()
# Checking that the market caps are positive
market_cap = (data['market_cap']>=0).all()
# Checking that each ID only shows up once
id = len(data['id'].unique()) == len(data['id'])
check = [price, volume_24h, market_cap, id]
if False not in check:
  print("There were no anomalies found in the data.")
else:
  columns = ['price', 'volume_24h', 'market_cap', 'id']
  for i in list(range(0,4)):
    if check[i] == False:
      print("There were anomalies found in the " + columns[i] + " column.\n")
```

In the above code, we are checking to see if all the prices, volume changes, and market caps were greater than or equal to 0 as we shouldn't see any negative values. We were also making sure that each ID only shows up once in the table to ensure that no entries were duplicated.

```python
# Checking for missing values
missing = data.isnull().sum().sum()
if missing == 0:
  print("There are no missing values in the dataset.")
else:
  print("The data contains " + str(missing) + " missing values.")
```
