```python
import json
import stomp
import atexit

class MessageListener(stomp.ConnectionListener):
    def __init__(self, topic_name):
        self.topic_name = topic_name

    def on_message(self, message):
        try:
            message_content = message.body.encode('utf-8')
            # Parse the JSON message
            json_message = json.loads(message_content)
            # Process the JSON message
            print(f"Received JSON message '{json_message}' from topic '{self.topic_name}'")
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON message: {e}")

class Subscriber:
    def __init__(self, topic_name):
        self.topic_name = topic_name
        self.conn = stomp.Connection([('localhost', 61613)])  # Replace with your broker's address and port
        self.conn.set_listener('', MessageListener(topic_name))
        self.conn.connect()
        self.conn.subscribe(destination=self.topic_name, id=1)

    def disconnect(self):
        self.conn.disconnect()


subscriber1 = Subscriber('/topic/New York')


try:
    import threading
    threading.Event().wait()
except KeyboardInterrupt:
    subscriber1.disconnect()

atexit.register(subscriber1.disconnect)
```
